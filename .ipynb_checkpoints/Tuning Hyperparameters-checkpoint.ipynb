## Nesse notebook avaliamos diferentes configurações de hiperparâmetros de cada classificador

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.grid_search import GridSearchCV
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
clean_data = pd.read_csv('data/clean_data.csv')
clean_data_gender = pd.read_csv('data/clean_data_gender.csv')
clean_data_alignment = pd.read_csv('data/clean_data_alignment.csv')
clean_data_publisher = pd.read_csv('data/clean_data_publisher.csv')
clean_data_grouped = pd.read_csv('data/clean_data_grouped.csv')
clean_data_grouped_gender = pd.read_csv('data/clean_data_grouped_gender.csv')
clean_data_grouped_alignment = pd.read_csv('data/clean_data_grouped_alignment.csv')
clean_data_grouped_publisher = pd.read_csv('data/clean_data_grouped_publisher.csv')
```

## Avaliação dos melhores  K números de vizinhos do classificador K-NN

```python
def knn_eval_hyperparameter(dataset, feature):
    """
        Avalia o K-NN buscando o melhor número de vizinhos de uma dada feature
    """
    y = dataset[feature]
    x = dataset.drop(feature, axis=1)
    
    knn = KNeighborsClassifier(n_neighbors=5)
    
    krange = list(range(1, 31))
    pgrid = dict(n_neighbors=krange)

    grid = GridSearchCV(knn, pgrid, cv=10, scoring='f1_micro')
    grid.fit(x, y)
    
    grid_mean_scores = [result.mean_validation_score for result in grid.grid_scores_]
    plt.plot(krange, grid_mean_scores)
    plt.suptitle('K-NN - Atributo ' + feature, fontsize=15)
    plt.xlabel('K')
    plt.ylabel('Score F1')
```

```python
knn_eval_hyperparameter(clean_data_grouped, "Flight")
```

```python
knn_eval_hyperparameter(clean_data_grouped, "Super Strength")
```

```python
knn_eval_hyperparameter(clean_data_grouped, "Accelerated Healing")
```

```python
knn_eval_hyperparameter(clean_data_grouped_alignment, "Alignment")
```

```python
knn_eval_hyperparameter(clean_data_grouped_publisher, "Publisher")
```

```python
knn_eval_hyperparameter(clean_data_grouped_gender, "Gender")
```

```python
knn_eval_hyperparameter(clean_data_grouped, "Stamina")
```
