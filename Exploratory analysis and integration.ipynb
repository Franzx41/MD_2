### Tipos de representaes para dados ausentes:
-  <b> NaN </b>
-  <b> -99 </b> para nmeros
-  <b> \- </b>  para categoricas

```python
# Todas as colunas com poderes tem pelo menos um valor true
num_poderes = poderes.select_dtypes(include=bool).sum(axis = 0)
num_poderes[num_poderes == 0]
```

```python
# Todos herois nesse dataset tem pelo menos um poder
num_poderes = poderes.select_dtypes(include=bool).sum(axis = 1)
num_poderes[num_poderes == 0]
```

```python
poderes.dtypes.unique()
```

```python
# No tem valores nulo, -99 ou - no dataset dos superpoderes
print(poderes.select_dtypes(include='O').columns)
poderes.loc[poderes['hero_names']=='-'].shape
```

```python
herois_nan = herois.replace(['-', -99], np.NaN)
```

```python
herois_nan.info()
```

### Personagens com mesmo nome

```python
# mesmo nome, mas poderes diferentes
poderes[poderes["hero_names"].str.startswith("Batman")]
```

```python
# Dois batman, com mesmo nome, mas altura e peso diferentes
herois_nan[herois_nan["hero_names"]=='Batman']
```

### Soluo:
##### Decidimos manter os herois com mesmo nome,  pois eles tem dados diferentes e dessa forma o consideramos pessoas diferentes

# Integrao dos dataset de caractersticas de herois e superpoderes:

```python
# devido ao fato que a maioria das features so as de poderes, 
# decidimos remover linhas dos herois sem dados no dataset de poderes fazendo um right join
data = herois_nan.merge(poderes, how='right', on=['hero_names'])
data.head()
```

```python
data.shape
```

```python
# no tem nenhum dado com todas os valores das colunas iguais
data.drop_duplicates().shape
```

```python
data.to_csv('data/merged_data.csv',index=False)
```
